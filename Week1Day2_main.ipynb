## 1. Variables and Basic Types

A **variable** is just a name that points to a piece of data. In Python, you don't have to declare what *type* of data it is ahead of time — Python figures that out for you based on what you assign.

The three basic types you'll use constantly:

| Type | What it is | Example |
|---|---|---|
| `int` | whole number | `42` |
| `float` | decimal number | `3.14` |
| `str` | text (string) | `"hello"` |

Let's create some variables describing a single weather observation at our research station.


````python
import sys

print("Python int:", sys.getsizeof(10), "bytes")
print("Python float:", sys.getsizeof(3.14), "bytes")
print("Python string:", sys.getsizeof("Hello"), "bytes")
````

## Type	Memory behavior
** int and str: variable size **

| Data Type | C (Typical Size) | Python (`sys.getsizeof()`) | Notes |
|-----------|-----------------:|---------------------------:|-------|
| `char`    | 1 byte           | `"A"` ≈ 50 bytes          | Python strings are objects. |
| `short`   | 2 bytes          | N/A                       | No separate `short` type in Python. |
| `int`     | 4 bytes          | `10` ≈ 28 bytes           | Variable-size object; can grow for large integers. |
| `float`   | 4 bytes          | `3.14` ≈ 24 bytes         | Fixed-size floating-point object. |
| `double`  | 8 bytes          | N/A                       | Python `float` is implemented as a C `double`. |
| `string`  | N/A              | `"Hello"` ≈ 54 bytes      | Size depends on the number of characters. |

````python
print(sys.getsizeof(10))         # Small int
print(sys.getsizeof(10**20))     # Bigger int
print(sys.getsizeof(10**100))    # Much bigger int

print(sys.getsizeof(3.14))
print(sys.getsizeof(123456789.79308))

print("'A'      :", sys.getsizeof("A"), "bytes")
print("'Hello'  :", sys.getsizeof("Hello"), "bytes")
print("'Hello World' :", sys.getsizeof("Hello World"), "bytes")
print("'A' * 100:", sys.getsizeof("A" * 100), "bytes")
````

### Converting between types

Sometimes data comes in the "wrong" type for what you want to do with it. A classic example: data read from a file often arrives as **strings**, even if it's secretly a number. If you don't convert it, math will break.


````python
# This is a common headache: a number stored as a string
raw_reading = "72.3"   # this came from, say, a text file or web form

print(type(raw_reading))

# This next line would cause an error if uncommented — you can't add a string and a float!
# print(raw_reading + 1.0)

# Fix it by converting types:
clean_reading = float(raw_reading)
print(type(clean_reading))
print(clean_reading + 1.0)
````

The three conversion functions you'll use most:

- `int(x)` → convert to integer (drops anything after the decimal point)
- `float(x)` → convert to decimal number
- `str(x)` → convert to text

```python
int("10")      # 10
float("10")    # 10.0
str(10)        # "10"
int(9.8)       # 9   (NOT rounded — just chopped off!)
```

## 2. Lists and Dictionaries

One observation isn't very interesting. Real data science means working with *many* observations. Two structures will carry almost all of your data in this course:

- **Lists** → ordered collections (good for "many readings of the same thing")
- **Dictionaries** → labeled collections (good for "many different pieces of info about one thing")


### Lists

A list stores items in order, and you access them by **position** (starting at 0, not 1!).


````python
# Temperature readings (°F) taken every 2 hours at our station today
temperatures = [64.0, 66.5, 70.2, 74.8, 77.1, 75.6, 71.3]

print(temperatures)
print("Number of readings:", len(temperatures))

# Indexing starts at 0
print("First reading:", temperatures[0])
print("Third reading:", temperatures[2])
print("Last reading:", temperatures[-1])   # negative index = count from the end
````

````python
# Lists are mutable — you can change, add, and remove items
temperatures.append(69.4)        # add a new reading to the end
print(temperatures)

temperatures[0] = 65.0           # fix a misrecorded first reading
print(temperatures)

print("Highest reading:", max(temperatures))
print("Lowest reading:", min(temperatures))
print("Average reading:", sum(temperatures) / len(temperatures))
````

### Dictionaries

A dictionary stores **key–value pairs**. Instead of looking something up by position, you look it up by a meaningful label (the *key*).


````python
# One observation, described with labels instead of position
observation = {
    "station_id": "RISE-01",
    "hour": 14,
    "temperature_f": 74.8,
    "humidity_pct": 52,
    "conditions": "partly cloudy"
}

print(observation)
print("Temperature:", observation["temperature_f"])
print("Conditions:", observation["conditions"])
````

````python
# Dictionaries are also mutable
observation["humidity_pct"] = 55     # update a value
observation["wind_mph"] = 8          # add a brand new key

print(observation)

# .keys() and .values() let you see what's available
print("Keys:", observation.keys())
print("Values:", observation.values())
````

### Lists *of* dictionaries — how real datasets are often shaped

This is one of the most important patterns in data science: a **list of dictionaries**, where each dictionary is one record (one row), and the keys are the columns.


````python
# A small "dataset" — three observations from our station today
station_data = [
    {"hour": 8,  "temperature_f": 64.0, "conditions": "clear"},
    {"hour": 14, "temperature_f": 77.1, "conditions": "partly cloudy"},
    {"hour": 20, "temperature_f": 68.3, "conditions": "clear"},
]

print(station_data)
print("Number of records:", len(station_data))
print("Second record:", station_data[1])
print("Temperature of second record:", station_data[1]["temperature_f"])
````

## 3. Logic and Repetition

Two things make programs powerful instead of just "a list of instructions executed once":

1. **Logic** (`if` / `elif` / `else`) — making decisions based on data
2. **Repetition** (`for` and `while` loops) — doing something many times without copy-pasting code

These two ideas are the backbone of almost every data-processing script you'll ever write.


### `if`, `elif`, `else`

Python checks conditions top to bottom and runs the **first** block that matches.


````python
temperature_f = 74.8

if temperature_f < 50:
    category = "cold"
elif temperature_f < 70:
    category = "mild"
elif temperature_f < 85:
    category = "warm"
else:
    category = "hot"

print(f"{temperature_f}°F is classified as: {category}")
````

A few important details:
- Only **one** branch runs, even if multiple conditions are technically true — Python stops at the first match.
- `else` is optional, and catches "everything else."
- You can use `and`, `or`, and `not` to combine conditions.


````python
humidity_pct = 55

if temperature_f > 70 and humidity_pct > 50:
    print("Feels muggy")
elif temperature_f > 70 and humidity_pct <= 50:
    print("Warm but dry")
else:
    print("Not particularly warm")
````

### `for` loops

A `for` loop runs a block of code once for **each item** in a sequence (like a list).


````python
temperatures = [64.0, 66.5, 70.2, 74.8, 77.1, 75.6, 71.3]

for temp in temperatures:
    print(temp)
````

````python
# Combine a for loop with if/elif/else to classify every reading
for temp in temperatures:
    if temp < 70:
        category = "mild"
    elif temp < 75:
        category = "warm"
    else:
        category = "hot"
    print(f"{temp}°F -> {category}")
````

###  Worked Example: counting with a loop

Loops are also how you build up summary statistics by hand — useful for understanding what's happening "under the hood" before we later use shortcuts like `sum()`.


````python
total = 0
count = 0

for temp in temperatures:
    total = total + temp   # running total
    count = count + 1      # running count

average = total / count
print(f"Total: {total}")
print(f"Count: {count}")
print(f"Average: {average:.2f}°F")   # :.2f formats to 2 decimal places
````

### `while` loops

A `while` loop repeats **as long as a condition is true**. Use this when you don't know in advance how many times you'll need to repeat something.


````python
# Simulate a sensor "warming up" until it reaches a stable reading
reading = 50.0
target = 65.0

while reading < target:
    reading += 3.0   # shorthand for: reading = reading + 3.0
    print(f"Current reading: {reading}")

print("Sensor stabilized!")
````

**Common bug:** if the condition in a `while` loop never becomes false, your loop runs forever (an "infinite loop"). Always make sure something inside the loop is moving you toward the stopping condition.

## 4. Helpful Python Patterns


These two tools solve problems you'll hit *constantly* once you start looping over real data.


### `enumerate` — looping with position tracking

Sometimes you need both the **item** and its **position** in the list. Instead of manually managing a counter, `enumerate` does it for you.


````python
temperatures = [64.0, 66.5, 70.2, 74.8, 77.1, 75.6, 71.3]

# The clunky way (avoid this):
i = 0
for temp in temperatures:
    print(f"Reading #{i}: {temp}")
    i += 1
````

````python
# The clean way, using enumerate:
for i, temp in enumerate(temperatures):
    print(f"Reading #{i}: {temp}")
````

````python
# enumerate is especially handy for flagging specific readings
for i, temp in enumerate(temperatures):
    print(f"  High reading at position {i}: {temp}°F")
````

### `zip` — looping over two lists together

`zip` pairs up items from two (or more) lists by position, so you can loop over them together.


````python
hours = [8, 10, 12, 14, 16, 18, 20]
temperatures = [64.0, 66.5, 70.2, 74.8, 77.1, 75.6, 71.3]

for hour, temp in zip(hours, temperatures):
    print(f"At {hour}:00, the temperature was {temp}°F")
````

### Worked Example: combining `zip` and `if`

This is exactly the kind of pattern you'll use to scan real datasets for points of interest.


````python
hours = [8, 10, 12, 14, 16, 18, 20]
temperatures = [64.0, 66.5, 70.2, 74.8, 77.1, 75.6, 71.3]

peak_hour = None
peak_temp = -1000   # start absurdly low so anything beats it

for hour, temp in zip(hours, temperatures):
    if temp > peak_temp:
        peak_temp = temp
        peak_hour = hour

print(f"Peak temperature was {peak_temp}°F at {peak_hour}:00")
````

## 5. Functions

A **function** is a reusable, named block of code. Instead of rewriting the same logic every time, you write it once and *call* it whenever you need it. This is one of the biggest steps toward writing clean, professional code.


````python
def classify_temperature(temp_f):
    """Classify a Fahrenheit temperature into a simple category."""
    if temp_f < 50:
        return "cold"
    elif temp_f < 70:
        return "mild"
    elif temp_f < 85:
        return "warm"
    else:
        return "hot"

# Calling the function
print(classify_temperature(45))
print(classify_temperature(74.8))
print(classify_temperature(90))
````

Anatomy of a function:

```python
def function_name(parameter1, parameter2):
    """Optional docstring explaining what it does."""
    # body — does some work
    return result   # sends a value back to whoever called it
```

- `def` starts the definition.
- Parameters (like `temp_f`) are inputs — placeholders filled in when you *call* the function.
- `return` sends a value back. A function with no `return` gives back `None`.


````python
# A function can take multiple parameters and combine logic + loops
def average_temperature(temp_list):
    """Return the average of a list of temperatures."""
    total = 0
    for temp in temp_list:
        total += temp
    return total / len(temp_list)

temperatures = [64.0, 66.5, 70.2, 74.8, 77.1, 75.6, 71.3]
avg = average_temperature(temperatures)
print(f"Average temperature: {avg:.2f}°F")
````

### Local vs. global variables

A variable created **inside** a function only exists inside that function — it's *local*. Variables created **outside** any function are *global*, and can be read (but shouldn't usually be modified) from inside functions.


````python
global_label = "Station RISE-01"   # global — exists everywhere in the notebook

def describe_reading(temp_f):
    local_summary = f"{global_label}: {temp_f}°F, classified as {classify_temperature(temp_f)}"
    # local_summary only exists *inside* this function
    return local_summary

print(describe_reading(77.1))

# This would cause an error if uncommented — local_summary doesn't exist out here:
# print(local_summary)
````

 **Why this matters:** keeping variables local to functions prevents one part of your code from accidentally messing up another part — a huge source of bugs in larger programs.

### Clean import patterns

There's more than one way to import, and the way you choose affects how readable your code is:

```python
import statistics                  # use as: statistics.mean(...)
import statistics as stats         # use as: stats.mean(...)   (a common alias)
from statistics import mean        # use as: mean(...)          (imports just one function)
```

**Best practice for this course:** prefer `import library_name` (or a well-known alias like `import numpy as np`) so it's always clear, when reading your code later, *where* a function came from.


````python
# A peek ahead: this is the kind of library you'll use constantly starting tomorrow
import random

random.seed(42)   # makes the "random" results repeatable — good for reproducibility!
simulated_reading = round(random.uniform(60, 80), 1)
print(f"Simulated temperature reading: {simulated_reading}°F")
````

 **Reproducibility tip:** setting a `seed` like `random.seed(42)` means anyone who runs your notebook gets the *exact same* "random" results. This matters a lot in research — it's part of what it means for work to be reproducible.
